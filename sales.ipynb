```python
nan_df = all_data[all_data.isna().any(axis=1)]
display(nan_df.head())

all_data = all_data.dropna(how='all')
all_data.head()

```

```python
# clean data find 'or' and delete
```

```python
all_data=all_data[all_data['Order Date'].str[0:2]!='Or']
all_data.head()
```

# What was the best month for sales and how much was earned this month?

```python
all_data['Month'] = pd.to_datetime(all_data['Order Date']).dt.month
all_data.head()

all_data.head()
```

```python
# convert columns to the correct type
```

```python
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype('int32')
all_data['Price Each']=pd.to_numeric(all_data['Price Each'])
all_data.head()
```

```python
all_data['Sales']=all_data['Quantity Ordered']*all_data['Price Each']
all_data.head()
```

```python
# using rank to find maximum
best=all_data.groupby(['Month']).sum().sort_values(by='Sales', ascending=False).reset_index()


best['rank']=best['Sales'].rank(method='min',ascending = False)


best[best['rank']==1].head()



```

```python
# using seaborn to visualize the data
import seaborn as sns
sns.barplot(x='Month',y='Sales',data=best)
```

# What city had the highest number of sales

```python
# taking city name and state code from purchase address
all_data['City']= all_data['Purchase Address'].apply(lambda x: x.split(',')[1] + " " + x.split(',')[2].split(' ')[1])
all_data.head()
```

```python
result=all_data.groupby('City').sum().sort_values(by='Sales',ascending= False).reset_index()
result
```

```python
# using rank to grab the highest sales
result['rank']=result['Sales'].rank(method='min',ascending=False)
result[result['rank']==1]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.xticks(rotation='vertical')

sns.barplot(x='City',y='Sales',data=result)
```

# What products are most often sold together?

```python
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))



df.head()
```

```python
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(50)
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df['Grouped']:
    row_list = row.split(',')
#     you count pers of two
    count.update(Counter(combinations(row_list, 2)))
count.most_common(10)
    
```
